```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y= x**2
```

```python
y
```

```python
plt.plot(x,y, 'r-')
# plt.show() # Printing of plots, needded if one is not using Jupyter
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.set_title("smaller plot")
axes1.set_title("larger plot")
axes2.plot(y,x)
```

## Subplots

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
```

## Figure Size and DPI 

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("my_picture.png",dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x Square')
ax.plot(x,x**3,label='x Cubed')
ax.set_title("Title")
ax.set_ylabel("Y")
ax.legend(loc=(0.1,0.1))
```

```python
fig.savefig("singlefig,png", dpi=200)
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple",linewidth=3,alpha=0.6,linestyle="-", marker="o", markersize=20, markerfacecolor="yellow", markeredgeWidth=3,markeredgecolor="black")
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="#FF8C00") # RBG Hex Code
```
