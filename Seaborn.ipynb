## Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="kde")
```

```python
sns.pairplot(tips,hue='sex',palette="coolwarm")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color="black")
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## Matrix Plot

```python
flights= sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap="coolwarm")
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap="coolwarm",linecolor='black',linewidth=1)
```

```python
sns.clustermap(fp,cmap="coolwarm",linecolor="black",linewidth=1,standard_scale=1)
```

## Grids

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

## Regression Plots

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="smoker",markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,col="day",hue="sex",aspect=0.6,size=8)
```

## Style and Color

```python
sns.set_style("ticks")
sns.countplot(x="sex",data=tips)
sns.despine()
sns.set_context("poster")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="coolwarm")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="seismic")
```
